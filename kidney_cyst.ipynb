```python
def calculate_iou(box1, box2):
    x1_max = max(box1[0], box2[0])
    y1_max = max(box1[1], box2[1])
    x2_min = min(box1[2], box2[2])
    y2_min = min(box1[3], box2[3])

    intersection_area = max(0, x2_min - x1_max) * max(0, y2_min - y1_max)

    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union_area = box1_area + box2_area - intersection_area

    return intersection_area / union_area
```

```python
def evaluate_detections(predictions, ground_truths, iou_threshold=0.5):
    tp = 0
    fp = 0
    fn = 0

    # Buat list untuk menyimpan apakah ground truth telah dipasangkan dengan prediksi
    matched_gt = [False] * len(ground_truths)

    for pred in predictions:
        best_iou = 0
        best_gt_index = -1

        for gt_index, gt in enumerate(ground_truths):
            iou = calculate_iou(pred, gt)
            if iou > best_iou:
                best_iou = iou
                best_gt_index = gt_index

        if best_iou >= iou_threshold:
            if not matched_gt[best_gt_index]:
                tp += 1
                matched_gt[best_gt_index] = True
            else:
                fp += 1
        else:
            fp += 1

    fn = matched_gt.count(False)

    return tp, fp, fn
```

```python
# Contoh bounding boxes (xmin, ymin, xmax, ymax)
predictions = [
    [100, 100, 200, 200],
    [150, 150, 250, 250]
]

ground_truths = [
    [120, 120, 220, 220],
    [300, 300, 400, 400]
]

tp, fp, fn = evaluate_detections(predictions, ground_truths)
print(f'TP: {tp}, FP: {fp}, FN: {fn}')
```
